## problem statement

a retail company "ABC private limited wants to understand the customer purchase behavior (specifically, purchase amount) against various products of different categories.they have shared purchase summary of various customers for selected high volume products from last month. the data set aslo contains customer demographics(age, gender, martial status, city_type ,stay_in_current_city),productdetails(product_id and productcategory)and total purchase_amount from last month.

now, they want to build a model to predict the purchase amount of customer against the products which will help them to create personalized offer of the customers against different products.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train=pd.read_csv("train.csv")
df_train.head()
```

```python
df_test=pd.read_csv("test.csv")
df_test.head()
```

```python
df=pd.merge(df_train,df_test,how="outer")
df.head()
```

basic

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(["User_ID"],axis=1,inplace=True)
```

```python
df.head()
```

data preprocessing

changing gender colomn replacing f,m with 0,1

```python
df["Gender"]=df["Gender"].map({"F":0,"M":1})
df.head()
```

handle categorial feature age

```python
df["Age"].value_counts()
```

```python
df["Age"]=df["Age"].map({"0-17":1,'18-25':2,'26-35':3,"36-45":4,"46-50":5,"50-55":6,"55+":7})
```

```python
df["Age"].value_counts()
```

```python
# another technique
# Import label encoder 
from sklearn import preprocessing 
  
# label_encoder object knows  
# how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
df['Age']= label_encoder.fit_transform(df['Age']) 
  
df['Age'].unique() 
```

```python
df["City_Category"].unique()
```

```python
df_city=pd.get_dummies(df["City_Category"],drop_first=True)
df_city
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```
