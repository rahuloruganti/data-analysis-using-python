seaborn

```python
import seaborn as sns
import pandas as pd

from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("apple_quality.csv")
```

```python

df.head()
```

```python
sns.lineplot(x="Crunchiness",y="Size", data=df)
```

```python
sns.lineplot(x="Crunchiness",y="Size", data=df,hue="Quality",style="Quality",markers=True)
```

barplot with seaborn


```python
sns.barplot(x="Quality",y="Size", data=df)
```

```python
sns.barplot(x="Quality",y="Size", data=df,hue="Sweetness")
```

```python
sns.barplot(x="Quality",y="Size", data=df,hue="Sweetness",color="orange")
```

scatterplot with seaborn

```python
sns.scatterplot(x="Crunchiness",y="Size",data=df,hue="Quality",style="Quality")
```

histogram or distplot with seaborn

```python
sns.displot(df["Size"])
```

```python
sns.kdeplot(y=df["Size"])
```

```python
sns.displot(y=df["Size"],color="red",bins=6)
plt.show()
```

jointplot

```python
sns.jointplot(x="Size",y="Sweetness",data=df,color="red",kind="reg")
```

boxplot

```python
sns.boxplot(x="Size",y="Quality",data=df)
```
