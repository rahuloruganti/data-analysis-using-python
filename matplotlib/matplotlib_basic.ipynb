matplotlib basic

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

create a numpy array using arange

```python
arr=np.arange(1,10)
arr
```

```python
arr1=arr*3
arr1
```

lineplot

with title, x-axis,y-axis

```python
plt.plot(arr,arr1)
plt.title("lineplot")
plt.xlabel("arr")
plt.ylabel("arr1")
plt.show()
```

change color ,linestyle and linewidth

```python
plt.plot(arr,arr1, color="g",linestyle="dotted",linewidth=2)
plt.title("lineplot")
plt.xlabel("arr")
plt.ylabel("arr1")
plt.show()
```

```python
arr2=arr*2
arr2
```

```python
plt.plot(arr,arr1,color="orange",linewidth=3,linestyle=":")
plt.plot(arr,arr2,color="g",linewidth=4,linestyle="dotted")
plt.title("arr1,arr2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

subplot 


```python
plt.subplot(2,3,1)
plt.plot(arr,arr1,color="orange",linewidth=3,linestyle=":")
plt.grid(True)
plt.subplot(2,3,2)
plt.plot(arr,arr2,color="g",linewidth=4,linestyle="dotted")
plt.title("arr1,arr2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

bar plot


```python
student={"nani":54,"dimpu":55,"linkedin":78}
value=student.values()
k=student.keys()
```

```python
plt.bar(k,value)
```

```python
plt.bar(k,value)
plt.title("maths")
plt.ylabel("score")
plt.xlabel("names")
plt.show()
```

horizontal bar plot

```python
plt.barh(list(k),value,color="red")
plt.show()
```

scatterplot

two numeric entities with same number of element

```python
x=[1,2,3,4,5,6,7,8,9]
y=[13,4,3,35,57,68,7,99,7]
z=[1,2,4,6,8,99,7,5,4]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,marker="*",c="green",s=20)
```

```python
plt.scatter(x,y,marker=".",c="yellow",s=200)
plt.scatter(x,z,marker="*",c="blue",s=100)
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker=".",c="yellow",s=200)
plt.subplot(1,2,2)
plt.scatter(x,z,marker="*",c="blue",s=100)
```
