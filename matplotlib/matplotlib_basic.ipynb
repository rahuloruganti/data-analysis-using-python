matplotlib basic

```python
import numpy as np
from matplotlib import pyplot as plt
```

create a numpy array using arange

```python
arr=np.arange(1,10)
arr
```

```python
arr1=arr*3
arr1
```

lineplot

with title, x-axis,y-axis

```python
plt.plot(arr,arr1)
plt.title("lineplot")
plt.xlabel("arr")
plt.ylabel("arr1")
plt.show()
```

change color ,linestyle and linewidth

```python
plt.plot(arr,arr1, color="g",linestyle="dotted",linewidth=2)
plt.title("lineplot")
plt.xlabel("arr")
plt.ylabel("arr1")
plt.show()
```

```python
arr2=arr*2
arr2
```

```python
plt.plot(arr,arr1,color="orange",linewidth=3,linestyle=":")
plt.plot(arr,arr2,color="g",linewidth=4,linestyle="dotted")
plt.title("arr1,arr2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

subplot 


```python
plt.subplot(2,3,1)
plt.plot(arr,arr1,color="orange",linewidth=3,linestyle=":")
plt.grid(True)
plt.subplot(2,3,2)
plt.plot(arr,arr2,color="g",linewidth=4,linestyle="dotted")
plt.title("arr1,arr2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```
