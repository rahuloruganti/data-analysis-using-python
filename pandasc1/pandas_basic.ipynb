pandas(panel data) for data analytics

```python
import pandas as pd
```

series object is one-dimensional labeled array

```python
s=pd.Series([1,2,3,4,5])
s
```

```python
type(s)
```

unlike numpy array this as labels and we can also change those labels

```python
s=pd.Series([1,2,3,4,5],index=["a","b","c","d","e"])
s
```

we can also create series with dictinaries

```python
s1=pd.Series({"a":1,"b":2,"c":3})
s1
```

indexing

same as list

```python
s[3]
```

basic math operations 

```python
# adding a scalar
s+2
```

```python
s+s1
```

```python
s*s1
```
